```python
# scatter plot for the whole data with average grouped by year
plt.scatter(df.trading_duration, df.avr_rev, color='blue', marker='s')
plt.xlabel('Trading Duration (in years)')
plt.ylabel('Avg Rev (in units of currency)')
plt.xlim(10,50)
plt.ylim(5000,300000)
plt.title('Trading Duration vs Avg Revenue')
m5,b=np.polyfit(df.trading_duration, df.avr_rev,1)
plt.plot(df.trading_duration,m5*df.trading_duration+b, color='orange')
```

```python
import seaborn as sns
```

```python
road1S=pd.read_sql_query('SELECT * FROM road1S', conn)
road1S.insert(1, 'trading_duration', 2014-road1S.YearOpened)
road1S.head()
```

```python
road1S.corr()
```

```python
# Specialty: Road
sns.set_style('whitegrid')
sns.color_palette()
sns.set_context('talk', font_scale=1.1)
sns.scatterplot(data=road1S, x='trading_duration', y='avg_rev', hue='number_shops', palette='Blues')
plt.xlabel('Trading Duration (years)   r=-0.313')
plt.ylabel('Avg Rev')
plt.title('Trading Duration vs Avg Revenue. Specialty: Road')
plt.legend(loc='upper left',fontsize='x-small', bbox_to_anchor=(0,-0.3), title='Number of Stors')
m,b=np.polyfit(road1S.trading_duration, road1S.avg_rev,1)
plt.plot(road1S.trading_duration,m*road1S.trading_duration+b, color='orange')

plt.show()
print(m)
```
